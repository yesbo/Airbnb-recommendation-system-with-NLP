```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
trainx = train.iloc[:, range(2,np.shape(train)[1])]
trainx.drop('review_scores_rating', axis = 1, inplace=True)
trainy = train['review_scores_rating']
```

```python
model = sklearn.ensemble.RandomForestRegressor(max_depth=10, n_estimators=200)
```

```python
model.fit(trainx, trainy)
```

```python
testx = test.iloc[:, range(2,np.shape(test)[1])]
testx.drop('review_scores_rating', axis = 1, inplace=True)
testy = test['review_scores_rating']
```

```python
predtest = model.predict(testx)
predtrain = model.predict(trainx)
```

```python
testerror = sklearn.metrics.mean_squared_error(y_pred=predtest, y_true=testy)
```

```python
math.sqrt(testerror)
```

```python
trainerror = sklearn.metrics.mean_squared_error(y_pred=predtrain, y_true=trainy)
```

```python
math.sqrt(trainerror)
```
